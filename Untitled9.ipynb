```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target function f(x1, x2)
def target_function(x1, x2):
    return 12 * (x1**2 + x2**2) - 1 - (1 / 100)

# Define the sigmoid activation function and its derivative
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Initialize the weights and biases for the network
def initialize_parameters(input_size, hidden_size, output_size):
    np.random.seed(42)  # For reproducibility
    W1 = np.random.randn(hidden_size, input_size) * 0.01
    b1 = np.zeros((hidden_size, 1))
    W2 = np.random.randn(output_size, hidden_size) * 0.01
    b2 = np.zeros((output_size, 1))
    return W1, b1, W2, b2

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(W1, X) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = Z2  # Linear activation for output layer (regression)
    return Z1, A1, Z2, A2

# Compute cost function (Mean Squared Error)
def compute_cost(A2, Y):
    m = Y.shape[1]
    cost = (1 / (2 * m)) * np.sum((A2 - Y) ** 2)
    return cost

# Backward propagation
def backward_propagation(X, Y, Z1, A1, A2, W2):
    m = X.shape[1]

    dZ2 = A2 - Y
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)

    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)

    return dW1, db1, dW2, db2

# Update weights and biases using gradient descent
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2

# Generate random data for training and testing
def generate_data(n_samples):
    X1 = np.random.uniform(-1, 1, n_samples)
    X2 = np.random.uniform(-1, 1, n_samples)
    X = np.vstack((X1, X2))
    Y = target_function(X1, X2).reshape(1, n_samples)
    return X, Y

# Train the network
def train_network(X, Y, hidden_size, learning_rate, num_iterations):
    input_size = X.shape[0]
    output_size = Y.shape[0]

    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

    costs = []

    for i in range(num_iterations):
        # Forward propagation
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

        # Compute cost
        cost = compute_cost(A2, Y)
        costs.append(cost)

        # Backward propagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, Z1, A1, A2, W2)

        # Update parameters
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost}")

    return W1, b1, W2, b2, costs

# Test the network
def test_network(X, Y, W1, b1, W2, b2):
    _, _, _, A2 = forward_propagation(X, W1, b1, W2, b2)
    test_cost = compute_cost(A2, Y)
    return test_cost

# Main script
if __name__ == "__main__":
    np.random.seed(42)

    # Generate training data (200 points)
    X_train, Y_train = generate_data(200)

    # Generate test data (100 points)
    X_test, Y_test = generate_data(100)

    # Define network architecture
    hidden_size = 10
    learning_rate = 0.1
    num_iterations = 5000

    # Train the network
    W1, b1, W2, b2, costs = train_network(X_train, Y_train, hidden_size, learning_rate, num_iterations)

    # Plot the cost over iterations
    plt.plot(costs)
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("Training Cost over Iterations")
    plt.show()

    # Calculate training and test errors
    train_cost = compute_cost(forward_propagation(X_train, W1, b1, W2, b2)[3], Y_train)
    test_cost = test_network(X_test, Y_test, W1, b1, W2, b2)

    print(f"Training Cost: {train_cost}")
    print(f"Test Cost: {test_cost}")
```
